<h3>Необов'язкові параметри команди print</h3>
За замовчуванням команда print() приймає кілька аргументів (параметрів), виводить їх через пробіл, після чого ставить перехід рядка. Цю поведінку можна змінити, використовуючи необов'язкові іменовані параметри sep (separator, роздільник) та end (закінчення).

Розглянемо наступний код:

```python
print('a', 'b', 'c')
print('d', 'e', 'f')
```

```python
print('a', 'b', 'c', sep='*')
print('d', 'e', 'f', sep='**')
```

При першому друку як рядок роздільника між аргументами команди print() встановлено рядок sep='*'.

При другому друку як рядок роздільника між аргументами команди print() встановлено рядок sep='**'.

Таким чином, необов'язковий параметр sep команди print() дозволяє встановити рядок, за допомогою якого будуть розділені аргументи під час друку.

Параметр end

Якщо перехід рядка робити не потрібно або потрібно вказати спеціальне закінчення, слід явно вказати значення для параметра end.

```python
print('a', 'b', 'c', end='@')
print('d', 'e', 'f', end='@@')
```

Після першого друку вставлений рядок @ замість перехіду рядка. Аналогічно, після завершення другого друку вставлено рядок @@.

Параметри sep та end можна використовувати разом. 

Розглянемо наступний код:

```python
print('a', 'b', 'c', sep='*', end='finish')
print('d', 'e', 'f', sep='**', end='^__^')
print('g', 'h', 'i', sep='+', end='%')
print('j', 'k', 'l', sep='-', end='#')
print('m', 'n', 'o', sep='/', end='!')
```

Виклик команди print() із порожніми дужками ставить перехід рядка.

Послідовність символів \n називається керуючою послідовністю і задає перехід рядка.

Значення за замовчуванням у параметрах sep і end:

sep=' ' # пробіл <br>
end='\n' # перехід рядка

Щоб прибрати всі додаткові символи, можна викликати командуprint() так:

```python
print('a', 'b', 'c', sep='', end='')
```

Якщо після виведення даних потрібно більше одного переходу рядка, необхідно використовувати наступний код:

```python
print('Python', end='\n\n\n')
print('Python')
```
