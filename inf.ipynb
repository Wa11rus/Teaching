Для виведення даних на екран використовується команда 

```python
print()
```

Усередині круглих дужок пишемо, що хочемо вивести на екран. Якщо це текст, обов'язково вказуємо його всередині лапок. Лапки можуть бути одинарними або подвійними. Тільки обов'язково ставимо однакові до та після тексту.

Наприклад, наступний код:

```python
print('Ми вивчаємо мову Python')
```

Те, що ми пишемо в круглих дужках команди print(), називається аргументами або параметрами команди.

Команда print() дозволяє вказувати кілька аргументів, у разі їх треба відокремлювати комами.

Наприклад, наступний код:

```python
print('Скоро я', 'програмуватиму', 'мовою', 'Python!')
```

Зверніть увагу, як розділювач при виведенні даних між аргументами команди використовується символ пробілу. За замовчуванням команда print() додає рівно одну прогалину між усіма своїми аргументами.

Для зчитування даних у Python використовується команда input().

```python
print('Как тебя зовут?')
name = input()
print('Привет,', name)
```

Предыдущий код можно переписать так:

```python
name = input('Как тебя зовут?')
print('Привет,', name)
```

Усі попередні програми, які ми писали, працювали з текстовими даними. Справді, команда input() зчитує рядок тексту. Однак у багатьох випадках нам потрібно працювати саме з числами. Щоб у Python створити змінну цілого типу даних, потрібно опустити лапки при оголошенні змінної. Розглянемо наступний код:

```python
num1 = 7            # num1 - це число
num2 = 10           # num2 - це число
num3 = num1 + num2  # num3 - це число
print(num3)
```

У мові Python, як і математики, над числами можна здійснювати 4 основні операції (+, −, *, /).

У математиці існує порядок виконання операцій, який визначає, які операції повинні виконуватися раніше за інших, навіть якщо у виразі вони написані правіше. Порядок виконання операцій у Python аналогічний порядку виконання операцій, які ви вивчали під час уроків математики.

Розглянемо наступний код:

```python
num1 = 2 + 3 * 4
num2 = (2 + 3) * 4
print(num1)
print(num2)
```

Перетворення рядка до цілого числа. Щоб перетворити рядок до цілого числа, ми використовуємо команду int(). Розглянемо наступний код:

```python
s = '1992'
year = int(s)
```

Змінна s має рядковий тип даних. За допомогою команди int() ми перетворили рядок до цілого числа і записали результат в змінну з ім'ям year. Ми використовуємо нову команду (точніше, функцію) – int(). Вона означає наступне: «Візьми те, що зазначено як аргумент у дужках, і перетвори це на ціле число». Таким чином, змінна рік має цілий тип даних, а змінна s – рядковий тип даних.

Напишемо програму, яка зчитує два цілих числа і виводить на екран їхню суму. 

```python
num1 = input()
num2 = input()
print(num1 + num2)
```

Python вважає, що в змінних num1 та num2 знаходиться текст, оскільки команда input() за замовчуванням зчитує саме текст. Для того, щоб явно вказати, що потрібно працювати зі змінними цілого типу, треба написати так:

```python
num1 = int(input())
num2 = int(input())
print(num1 + num2)
```

Умовний оператор if-else

Ми познайомилися з базовими будівельними блоками програм, навчилися писати програми, що забезпечують введення, обробку та виведення даних. Більше того, вміємо працювати з рядками та числами, як ми робимо це у математиці. Тепер навчимося керувати ходом виконання програми.

Програми повинні вміти виконувати різні дії, залежно від введених даних. Для прийняття рішення програма перевіряє, чи істинно чи хибно певна умова.

У Python існує кілька способів перевірки, і в кожному випадку можливі два результати: істина (True) або брехня (False).

У Python перевірка умов здійснюється за допомогою ключового слова if.

Розглянемо таку програму:

```python
answer = input('Яку мову програмування ми вивчаємо?')
if answer == 'Python':
    print('Python - чудова мова!')
```

Попередня програма виводить текст у випадку, якщо умова є істинною. Але якщо умова помилкова, то програма нічого не виводить. Для того, щоб забезпечити можливість виконувати будь-що, якщо умова виявилася помилковою, ми використовуємо ключове слово else.

```python
answer = input('Яку мову програмування ми вивчаємо?')
if answer == 'Python':
    print('Python - чудова мова!')
else:
     print('Не зовсім так!')
```
